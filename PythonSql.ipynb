```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
# create_engine = motor que fara meu banco de dados rodar // Medatada = CLasso do banco de dados
```

```python
banco_dados = create_engine('sqlite:///memory:')
#sqllite, ja ve instalado // Memory informa que esotu usando memoria local.
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url) #Leitura do CSV
dados.head()
```

```python
#dados.to_sql('clientes', banco_dados, index = False) # nome do banco, motor de criação
```

```python
var_inspector = inspect(banco_dados)#verificação do banco de dados
```

```python
print(var_inspector.get_table_names()) # printa a tabela
```

```python
# consultas SQl dentro da Tabela
```

```python
var_query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
empregados = pd.read_sql(var_query, banco_dados) # junto o motor do banco de dados.
empregados.head()
```

```python
#empregados.to_sql('empregado', con=banco_dados) #junto ao con+ motor assim savei essa nova tabela
```

```python
pd.read_sql_table('empregado', banco_dados)# lendo a nova tabela completa
```

```python
pd.read_sql_table('empregado', banco_dados, columns=['ID_Cliente','Grau_escolaridade','Categoria_de_renda'])#com algumas colunas especificas
```

```python
# Atualizando bando de dados.
```

```python
query = '  DELETE FROM clientes WHERE ID_Cliente = 5008804'
with banco_dados.connect() as conn:
  conn.execute(text(query))
  conn.commit()
  # delete realizado
```
